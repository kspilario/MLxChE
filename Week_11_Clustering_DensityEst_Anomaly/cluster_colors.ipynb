# KMeans: Extract Color Palettes from Images

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans

image = mpimg.imread('sound_of_music.jpg')
plt.imshow(image)
plt.title('Original Image')
plt.show()
w, h, d = image.shape
data = np.reshape(image, (w*h, d))
n_colors = 10
model = KMeans(n_clusters=n_colors, 
               random_state=42,
               n_init=10).fit(data)
palette = model.cluster_centers_.astype(np.uint8)   # Use this if values are [0...255]
#palette = model.cluster_centers_                   # Use this if values are [0...1]
plt.imshow([palette])
plt.show()

print('Pixel values:')
print(palette)
```
