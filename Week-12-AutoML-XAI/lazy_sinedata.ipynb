# Lazy Regressor: Sine Data Set
Reference: https://pypi.org/project/lazypredict/

```python
import piplite
await piplite.install('lazypredict')
import lazypredict
import numpy as np
import matplotlib.pyplot as plt
from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split

np.random.seed(0)
x = np.linspace(0,24,300).reshape(-1,1)
y = np.sin(x) + 0.5*np.random.rand(x.size,1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=0)
y_train, y_test = np.ravel(y_train), np.ravel(y_test)

def plotData(X_train,X_test,y_train,y_test):
    plt.scatter(X_train,y_train,color='b',alpha=0.3,label='Training Data')
    plt.scatter(X_test,y_test,color='m',alpha=0.3,label='Test Data')
    plt.ylim([-2.5, 2.5])
    plt.grid()
    
plotData(X_train, X_test, y_train, y_test)
plt.legend()
plt.show()
```

```python
mods = lazypredict.Supervised.REGRESSORS
model_list = mods[:30] + mods[35:]
lazypredict.Supervised.REGRESSORS = model_list

reg = LazyRegressor(verbose=1, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

gpr = GaussianProcessRegressor().fit(X_train,y_train)
X_all = np.linspace(0,24,200).reshape(-1,1)
y_all = gpr.predict(X_all)
plotData(X_train, X_test, y_train, y_test)
plt.plot(X_all,y_all,'r',label='GPR mean')
plt.legend()
plt.show()
```
